# Homework 0

While we'll be using remote Google Colab or local Jupyter servers to execute our code, keeping track of changes and such is a separate thing.
We'll use `git` for storing all the code for the course, both mine and yours.

First thing to do is to navigate to https://github.com and get an account there, if you don't have one already.

Next, you should fork the PHYS 152 repo that I have set up: https://github.com/JasonNielsen/phys152-2026.  You can read more about forking in github here: https://docs.github.com/en/pull-requests/collaborating-with-pull-requests/working-with-forks/fork-a-repo.

You should follow the instructions to create and update a fork.
You will not need to do merge/pull requests or push changes upstream in this course.

Please name your fork `phys152-2026`, and not something else.
This will make life much easier for the instructors!

Now, from the new forked project: click on the "Manage/Members" link on the panel, and add me (`JasonNielsen` on github) to your project.
Eventually you will also need to add the grader.
If you notice other people have access to your fork, it's probably because they have inherited access to you fork through their own fork.
You can (and should) remove their permissions.

When you're done, submit a link to your PHYS 152 git repository in Canvas.

That's it for Homework 0!
